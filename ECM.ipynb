```
import matplotlib.pyplot as plt 
```

```
x=[0, 20, 30, 40, 50, 60]
y=[100, 150,200,180,250,230 ]
y_p = []
```

# Regresion lineal
image.png

```
def sum_xy(x, y):
  sum_xy = 0
  for index, _ in enumerate(x):
    sum_xy += x[index]*y[index]
  return sum_xy
```

```
def pendiente(x, y):
  m_num = len(x)*sum_xy(x, y) - sum(x)*sum(y)
  m_den = len(x) * sum([x_val ** 2 for x_val in x])  - sum(x) ** 2
  return m_num/m_den
```

```
def interseccion(x,y):
  return (sum(y) - pendiente(x, y) * sum(x))/len(x)
```

```
# Funcion lineal
def f(x, m, b):
  return m * x + b
```

```
# calculo de y_prima
for x_val in x:
  y_p.append(round(f(x_val, pendiente(x, y), interseccion(x, y)), 2))
print(y_p)
```

# Grafica

```
fig,aux=plt.subplots()
aux.plot(x,y_p) # linea
aux.scatter(x,y,color='r') # puntos
aux.grid()
fig.show()
```

# Error cuadratico medio

```
def ecm(y, yp):
  return len(y) * sum([yp[idx] - y[idx] for idx, _ in enumerate(y)])
```

```
# calculo del ecm
print(f'El ECM es {ecm(y, y_p)}')
```
